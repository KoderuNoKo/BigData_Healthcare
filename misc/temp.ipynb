# Test reading data from google bigquery

```python
from google.cloud import bigquery
import pandas as pd

client = bigquery.Client(project='bigdata-healthcare')

query = """
    SELECT 
        subject_id, 
        gender, 
        anchor_age
    FROM 
        `physionet-data.mimiciv_3_1_hosp.patients`
    LIMIT 10
"""

query_job = client.query(query) 
df = query_job.to_dataframe()

print(df)
```
